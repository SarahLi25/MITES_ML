## Problem 1

Write a function that splits a string on a single character. This function should behave identically to the Python `split` method that can be called on strings. You can assume we will only split on a single character. The challenge is to, of course, avoid relying on the `split` method!

```python
# here is a string that you can test your code on!
text = 'This is some example text! You can edit it as needed for testing purposes!'
```

```python
"""
    This function splits a string into a list of strings based on a separator.

    My original solution.

    Args:
        string (str): A string to be split.
        separator (str): A string that indicates where to split the string.

    Returns:
        list[str]: A list of strings that were separated by the separator.
"""
def split_string(string, separator):
    # holds the split parts
    result = []   # list hold the strings
    current= ""
    for char in string:             # if the seperator is found = append the current string to the result then reset current
        if char == separator:
            result.append(current)
            current = ""
        else:                       # if the character is not the separator = add it to the current string
            current += char
    result.append(current)

    return result
```

```python
# tests - DO NOT MODIFY! - run this cell to see if your code works!
chars = 'rstlneRSTLNE!.,? '

for c in chars:
    result = split_string(text, c)
    true = text.split(c)
    assert result == true, f'Error:\nResult: {result}\nTrue: {true}'

print('All tests passed!')
```

## Problem 2

Write a function that finds the sum of the first $n$ Fibonacci numbers. The first two Fibonacci numbers are 1, and the next number is the sum of the previous two. Here is how the first few numbers are calculated:

The first number is **1** (sum so far: 1)

The second number is **1** (sum so far: 1 + 1 = 2)

*Now, we add the first two numbers to get the third:*

The third number is **2** = 1 + 1 (sum so far: 1 + 1 + 2 = 4)

*Now, we add the second and third numbers to get the fourth:*

The fourth number is **3** = 1 + 2 (sum so far: 1 + 1 + 2 + 3 = 7)

*Now, we add the third and fourth numbers to get the fifth:*

The fifth number is **5** = 2 + 3 (sum so far: 1 + 1 + 2 + 3 + 5 = 12)

*And so on...*

The sixth number is **8** = 3 + 5 (sum so far: 1 + 1 + 2 + 3 + 5 + 8 = 20)

The seventh number is **13** = 5 + 8 (sum so far: 1 + 1 + 2 + 3 + 5 + 8 + 13 = 33)

...

```python
""" This function calculates the sum of the first n Fibonacci numbers.

    My original solution.

    Args:
        n (int): The number of Fibonacci numbers to sum.

    Returns:
        int: The sum of the first n Fibonacci numbers.
"""

def sum_fib(n):
  if n == 0:  # no num = no sum
    return 0

  if n == 1:  # only if 1st fib num
    return 1

  # the 1st AND 2nd are both 1
  a = 1
  b = 1

  total = 2 # sum starts with 2

  for i in range (n-2): # sub the 2 from the sum of num ur trying to find
    next_num = a+b
    total += next_num   # add to running total

    a=b   #update the previous two nums
    b=next_num

  return total
```

```python
# tests - DO NOT MODIFY! - run this cell to see if your code works!
correct = (0, 1, 2, 4, 7, 12, 20, 33)

for i in range(len(correct)):
    assert sum_fib(i) == correct[i], f'Error:\nResult: {sum_fib(i)}\nCorrect: {correct[i]}'

print('All tests passed!')
```
